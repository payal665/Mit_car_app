```python
### import pandas as pd

df = pd.read_csv('Car Dataset Processed.csv')
df.head()

# insurance_validity
d1 = {
    'comprehensive':0,
    'Third Party insurance':1,
    'Zero Dep':2,
    'Not Available':3,
    'Third Party':1
}

df['insurance_validity'] = df['insurance_validity'].map(d1)


# fuel_type
d2 = {
    'Petrol':0,
    'Diesel':1,
    'CNG':2
}

df['fuel_type'] = df['fuel_type'].map(d2)


# ownsership
d4 = {
    'First Owner':0,
    'Second Owner':1,
    'Third Owner':2,
    'Fourth Owner':3,
    'Fifth Owner':4
}

df['ownsership'] = df['ownsership'].map(d4)


# transmission
d3 = {
    'Manual':0,
    'Automatic':1
}

df['transmission'] = df['transmission'].map(d3)

df = df.dropna()


# Features
X = df[['insurance_validity', 'fuel_type', 'kms_driven', 'ownsership', 'transmission']]
y = df['price(in lakhs)']


# Models
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR


models = {
    'LinReg': LinearRegression(),
    'KNN-3': KNeighborsRegressor(n_neighbors=3),
    'SVM_Poly': SVR(kernel='poly')
}


for name, model in models.items():
    model.fit(X, y)
    sc = model.score(X, y)
    print(name, sc)


# Final Model
final_model = models['KNN-3']


# Input
insurance_validity = input('Enter insurance_validity:')
fuel_type = input('Enter fuel_type:')
kms_driven = input('Enter kms_driven:')
ownsership = input('Ownsership:')
transmission = input('Transmission type:')


insurance_validity = d1[insurance_validity]
fuel_type = d2[fuel_type]
transmission = d3[transmission]
ownsership = d4[ownsership]
kms_driven = int(kms_driven)


test = [[insurance_validity, fuel_type, kms_driven, ownsership, transmission]]

yp = final_model.predict(test)[0]

print("Predicted car price is {} lakhs".format(yp))
```

```python
import pickle 
pickle.dump(final_model,open('final_model.pkl','wb'))
```
